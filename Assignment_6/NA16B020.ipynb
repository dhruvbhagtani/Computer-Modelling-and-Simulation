This means that the rank of $D^2$ is N-1. Following this, the rank of $D^K$ will be (N+1-K). 

Therefore, if we have $D^(N+1)$, its rank will be (N+1-(N+1)) = 0. This means that the rank of the matrix will be zero, and the power henceforth, will be zero. The 2nd norm of the (N+1)th derivative of the Chebyshev differentation matrix is given by the following code.


```python
import scipy as sp
import numpy as np
import matplotlib.pylab as plt
def Cheb(N):
    D = np.zeros(shape=(N+1,N+1))
    D[0,0] = (2*N**2 + 1)/6;
    D[N,N] = -D[0,0];
    for i in range (1,N):
        x = sp.cos(sp.pi*i/N);
        for j in range(1,N):
            y = sp.cos(sp.pi*j/N);
            if (i == j):
                D[i,i] = -x/(2*(1-x**2))
            else:
                D[i,j] = ((-1)**(i+j))/(x-y)
    for i in range(1,N):
        x = sp.cos(sp.pi*i/N);
        D[0,i] = (2*(-1)**i)/(1-x)
        D[N,i] = (-2*(-1)**(N+i))/(1+x)
        D[i,0] = (-0.5*(-1)**i)/(1-x)
        D[i,N] = (0.5 * (-1)**(N+i))/(1+x)
   
    D[0,N] = (0.5)*((-1)**N)
    D[N,0] = -(0.5)*((-1)**N)
    
    temp = D;
    for i in range (1,N+2):
        temp = sp.matmul(temp,D)
    nor = np.linalg.norm(temp);
    if (N == 5 or N == 20):
        print(nor);
    return nor;

max = 25;
norm = sp.zeros(max-2);
N = sp.zeros(max-2);
for j in range(2,max):
    norm[j-2] = Cheb(j);
    N[j-2] = j;

plt.semilogy(N,norm)
plt.show();
    
```

This is a semilog plot of N vs the norm of the (N+1)th differentation of D. As can be seen, it increases exponentially with N. For N = 5, the norm is almost 0 and for N = 20, it is of the order $10^{21}$.
