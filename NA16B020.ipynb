a. Part 1:
    Fourier transform of any function is given by:
    \begin{equation}
    \hat f(k) = \int_{-\infty}^{\infty}e^{-ikx}f(x)dx        
    \end{equation}
    for all $k \in \mathbb{R}$
    
If $f(x) = u(x) + v(x)$, then we can write the above equation as 
    \begin{equation}
    \hat f(k) = \int_{-\infty}^{\infty}e^{-ikx}(u(x) + v(x))dx        
    \end{equation}
    \begin{equation}
    \hat f(k) = \int_{-\infty}^{\infty}e^{-ikx}u(x)dx + \int_{-\infty}^{\infty}e^{-ikx}v(x)dx        
    \end{equation}
    \begin{equation}
    \hat f(k) = \hat u(k) + \hat v(k)        
    \end{equation}
    
   Part 2:
   Let the function $f(x) = cu(x)$, then
   
   \begin{equation}
   \hat f(k) = \int_{-\infty}^{\infty}e^{-ikx}(cu(x))dx       
   \end{equation}
    
   Since c is a constant, we can take it out from the RHS and write it as:
    
   \begin{equation}
   \hat f(k) = c\int_{-\infty}^{\infty}e^{-ikx}(u(x))dx   
   \end{equation}
    
Therefore, both are equal.

Q4. The sinc function can be represented as:

\begin{equation}
S_h(x) = \frac{sin(\frac{\pi x}{h})}{\frac{\pi x}{h}}
\end{equation}

Differentiating it wrt x, we get

\begin{equation}
S_h'(x) = \frac{cos(\frac{\pi x}{h})}{\frac{\pi x}{h}} \frac{\pi}{h} - \frac{sin(\frac{\pi x}{h})}{\frac{\pi x^2}{h}}
\end{equation}

Since we are considering the discrete number line, where $x_j = jh$, placing it in the above equation,

\begin{equation}
S_h'(x_j) = \frac{cos(\pi j)}{jh} - \frac{sin(\pi j)}{\pi j^2 h}
\end{equation}

For $sin(\pi j))$, where j is an integer, the value is zero.

\begin{equation}
S_h'(x_j) = \frac{(-1)^j}{jh}
\end{equation}

only when j is not equal to zero.

When j = 0, the value is not defined and we take the limit near x = 0.

\begin{equation}
S_h'(x) = \frac{cos(\frac{\pi x}{h})}{\frac{\pi x}{h}} \frac{\pi}{h} - \frac{sin(\frac{\pi x}{h})}{\frac{\pi x^2}{h}}
\end{equation}

\begin{equation}
S_h'(x) = \frac{cos(\frac{\pi \Delta x}{h})}{\Delta x} - \frac{sin(\frac{\pi \Delta x}{h})}{\frac{\pi \Delta x^2}{h}}
\end{equation}

\begin{equation}
S_h'(x) = \frac{1}{\Delta x} - \frac{1}{\Delta x} = 0
\end{equation}


$S_h'(x)$ can be written as: 

\begin{equation}
S_h'(x) = \frac{cos(\frac{\pi x}{h})}{x} - \frac{sin(\frac{\pi x}{h})}{x^2}\frac{h}{\pi}
\end{equation}

Differentiating the function wrt time, we get

\begin{equation}
S_h''(x) = -\frac{cos(\frac{\pi x}{h})}{x^2} - \frac{sin(\frac{\pi x}{h})}{x}\frac{\pi}{h} + 2\frac{sin(\frac{\pi x}{h})}{x^3}\frac{h}{\pi} - \frac{cos(\frac{\pi x}{h})}{x^2}
\end{equation}

When j is not equal to zero, we can write the equation as :

\begin{equation}
S_h''(x) = -\frac{cos(\pi j)}{(jh)^2} - \frac{sin(\pi j)}{jh}\frac{\pi}{h} + 2\frac{sin(\pi j)}{(jh)^3}\frac{h}{\pi} - \frac{cos(\pi j)}{(jh)^2}
\end{equation}

From similar ideology, $sin(j\pi)$ is zero. Only the cos term remains, and we get,

\begin{equation}
S_h''(x) = -2\frac{(-1)^j}{(jh)^2}
\end{equation}

When j = 0, the function is not defined, and by putting limits as $x = \Delta x$, we get

\begin{equation}
S_h''(\Delta x) = -2\frac{cos(\frac{\pi \Delta x}{h})}{(\Delta x^2)} - \frac{sin(\frac{\pi \Delta x}{h})}{\Delta x}\frac{\pi}{h} + 2\frac{sin(\frac{\pi \Delta x}{h})}{\Delta x^3}\frac{h}{\pi}
\end{equation}

\begin{equation}
S_h''(\Delta x) = \frac{-2}{\Delta x^2} - \frac{sin(\frac{\pi \Delta x}{h})}{\Delta x}\frac{\pi}{h} + \frac{2}{\Delta x^2} - \frac{\pi^2}{h^2}
\end{equation}

\begin{equation}
S_h''(\Delta x) = \frac{-2}{\Delta x^2} - \frac{\pi^2}{h^2} + \frac{2}{\Delta x^2} - \frac{1}{3}\frac{\pi^2}{h^2}
\end{equation}

\begin{equation}
S_h''(\Delta x) = - \frac{4}{3}\frac{\pi^2}{h^2}
\end{equation}

Q5. For the central function, the multiplication happens like this:

\begin{gather}
D_N = 
\begin{bmatrix}
0&1&\ddots & & &\frac{1}{2} & -1\\
-1&0&\ddots & \frac{-1}{2} & &\\
 & & \ddots & 1\\
 & & \ddots & 0 & &\\
 &  & & \vdots \\
\end{bmatrix}
\end{gather}


Now, when we need to find the answer, we need to multiply $D_N$ by itself. Each element of the D_N^2 matrix is an infinite series sum, which is given by:

\begin{equation}
D_N(centre, centre) = -\frac{1}{n}\frac{1}{n} - \frac{1}{n-1}\frac{1}{n-1} + \dots + 0 -\frac{1}{2}\frac{1}{2} \dots
\end{equation}

This is actually $\frac{-\pi^2}{6}$. For the other terms, let's take an example of a row above the centre and the centre column. We get the series as:

\begin{equation}
D_N(centre-1, centre) = \frac{1}{n-1}\frac{1}{n} + \frac{1}{n-1}\frac{1}{n-2} + \dots + 0 + \frac{1}{1}\frac{1}{2} \dots
\end{equation}

which can be rewritten as:

\begin{equation}
2\sum_{n=1}^{\infty}\frac{1}{(2n)(2n+1)}
\end{equation}

This is equal to 1, as shown.

We could do that because of the following reason by Euler:

\begin{equation}
\frac{sin(x)}{x} = 1 - \frac{x^2}{3!} + \frac{x^4}{5} + \dots
\end{equation}

Using the Weierstrauss factorization theorem,

\begin{equation}
\frac{sin(x)}{x} = (1 - \frac{x}{\pi})(1+\frac{x}{\pi})(1-\frac{x}{2\pi})(1+\frac{x}{2\pi})(1-\frac{x}{3\pi})(1+\frac{x}{3\pi}) + \dots
\end{equation}

Collecting all the x^2 terms, we get

\begin{equation}
-\big(\frac{1}{\pi^2} + \frac{1}{4\pi^2} + \frac{1}{9\pi^2} + \dots) = -\frac{1}{\pi^2}\sum^{\infty}_{n=1}\frac{1}{n^2}
\end{equation}

Equating $x^2$ terms on both the sides, we get

\begin{equation}
\frac{-1}{6} = \frac{-1}{\pi^2}\sum^{\infty}_{n=1}\frac{1}{n^2}
\end{equation}

This can be written as:

\begin{equation}
\frac{-\pi^2}{6} = \sum^{\infty}_{n=1}\frac{1}{n^2}
\end{equation}

The reimann zeta function is written as:
\begin{equation}
\zeta(r) = \sum^{\infty}_{n=1}\frac{1}{n^r}
\end{equation}

This is why $\zeta(2)$ gives us the desired proof.

```python
# Band limited interpolation
import scipy as sp;
import numpy as np
import matplotlib.pyplot as plt
h=0.5**(3);
xmax=10;
N=2*xmax//h;
x=sp.linspace(-xmax,xmax,N+1); # computational grid
xx=sp.linspace(-xmax-h/20,xmax+h/20,10*(N+1)); # plotting grid
error = np.zeros((6,3));
for k in range(0,6):
    h = 0.5**(k-1)
    N=2*xmax//h;
    x=sp.linspace(-xmax,xmax,N+1); # computational grid
    xx=sp.linspace(-xmax-h/20,xmax+h/20,10*(N+1)); # plotting grid
    for i in range(1,4):
        plt.subplot(3,1,i);
        if (i==1):
            v=(x==0);
            v1=(xx==0);
        if (i==2):
            v=(sp.absolute(x)<=3.0);
            v1=(sp.absolute(xx)<=3.0);
        if (i==3):
            v=(sp.maximum(0,1-sp.absolute(x)/3.0));
            v1=(sp.maximum(0,1-sp.absolute(xx)/3.0));
        plt.plot(x,v,'ko');
        p = sp.zeros(sp.size(xx));
        for j in range (0,sp.size(x)):
            y = (xx-x[j])/h;
            p = p+v[j]*sp.sin(sp.pi*y)/(sp.pi*y);      
        plt.plot(xx,p,'r-');
        plt.axis([-11, 11, -0.25, 1.25]);
        error[k,i-1] = sp.amax(v1-p);
print((error));
#h1 = 1/sp.size(xx);
plt.rcParams["figure.figsize"] = [16,6]
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5);
plt.show();

for i in range(1,6):
    h = 0.5**(i-1)
    plt.loglog(1/h,error[i,1],'+')
plt.xlabel('Size of h');
plt.ylabel('Error');
plt.title('Box Function');
plt.show()

for i in range(1,6):
    h = 0.5**(i-1)
    plt.loglog(1/h,error[i,2],'*')
plt.xlabel('Size of h');
plt.ylabel('Error');
plt.title('Hat Function');
plt.show();
```

The values of error is constant for the box function, whereas for the hat function, it reduces by half as the number of grid points increases by 2.

```python
import scipy as sp;
import matplotlib.pyplot as plt
N = 2^4;
h = 2*sp.pi/N;
k = 2*sp.arange(3,16);
g2 = sp.zeros(sp.size(k));
g4 = sp.zeros(sp.size(k));
ginf = sp.zeros(sp.size(k));
for n in range(0,sp.size(k)):
    g2[n] = sp.sin(n*h)/h;
    g4[n] = ((-1/(6*h)) * sp.sin(2*n*h) + (4/(3*h)) * sp.sin(n*h));
    ginf[n] = n

import matplotlib.pyplot as plt
plt.loglog(k,g2,'+', label='2nd Order')
plt.loglog(k,g4,'^', label='4th Order')
plt.loglog(k,ginf,'*',label = 'inf')
plt.xlabel('Number of points');
plt.ylabel('Value of the function');
plt.legend();
plt.title('Fourth order convergence of the finite difference');
plt.show()
```
