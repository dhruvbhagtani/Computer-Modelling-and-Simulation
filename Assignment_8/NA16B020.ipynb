Q1.
a. Runge Kutta second order method

$k1 = \Delta t f(y_n,t_n)$

$k_2 = \Delta t f(y_n + \beta k_1, t_n + \alpha h)$

$y_{n+1} = y_n + \gamma_1 k_1 + \gamma_2 k_2$

where we need to evaluate the following to achieve the maximum accuracy.

Taylor expansion for $y_{n+1}$ is given as:

\begin{equation}
    y(t_{n+1}) = y(t_n) + (\Delta t) \frac{dy}{dt}\Big \vert_{t_n} + \frac{(\Delta t)^2}{2}\frac{d^2y}{dt^2}\Big \vert_{t_n} + O((\Delta t)^3)
\end{equation}

We know that $\frac{dy}{dt}$ = f(y,t), so we can write the third term of the above equation as follows:

\begin{equation}
    \frac{d^2y}{dt^2} = \frac{df(y,t)}{dt} = \frac{\partial f}{\partial t} + \frac{\partial f}{\partial y}\frac{dy}{dt} = \frac{\partial f}{\partial t} + \frac{\partial f}{\partial y}f
\end{equation}

Substituting this in the first equation, we get

\begin{equation}
    y(t_{n+1}) = y(t_n) + (\Delta t) \frac{dy}{dt}\Big \vert_{t_n} + \frac{(\Delta t)^2}{2}\Big(\frac{\partial f}{\partial t} + \frac{\partial f}{\partial y}f\Big)\Big \vert_{(y_n, t_n)} + O((\Delta t)^3)
\end{equation}

Expanding 

\begin{equation}
    k_2 = (\Delta t)f(y_n + \beta k_1, t_n + \alpha \Delta t)
\end{equation}

we get

\begin{equation}
    k_2 = (\Delta t)(f(y_n,t_n) + \alpha \Delta t \frac{\partial f}{\partial t}(y_n,t_n) + \beta k_1 \frac{\partial f}{\partial y}(y_n,t_n)) + O((\Delta)^3)
\end{equation}

Substituting this value of $k_2$ in the following equation

\begin{equation}
    y_{n+1} = y_n + \gamma_1 k_1 + \gamma_2 k_2
\end{equation}

we get

\begin{equation}
    y_{n+1} = y_n + \gamma_1 \Delta t f(y_n,t_n) + \gamma_2((\Delta t)(f(y_n,t_n) + \alpha \Delta t \frac{\partial f}{\partial t}(y_n,t_n) + \beta k_1 \frac{\partial f}{\partial y}(y_n,t_n))) + O((\Delta)^3)
\end{equation}

\begin{equation}
    y_{n+1} = y_n + (\gamma_1 + \gamma_2) \Delta t f(y_n,t_n) + \gamma_2(\Delta t^2) (\alpha \frac{\partial f}{\partial t} + \beta f \frac{\partial f}{\partial y})(y_n,t_n) + O((\Delta t)^3)
\end{equation}

Upon comparing, we get

$\gamma_1 + \gamma_2 = 1$

$\beta \gamma_2 = \frac{1}{2}$

$\alpha \gamma_2 = \frac{1}{2}$

These are the relations we obtain between the coefficients. Since there are 4 unknowns and 3 equations, they can't be uniquely determined from the above equations only.

Region of stability in the x-y plane for RK2 and RK4
A single step ODE is written in the form as:

\begin{equation}
y_{n+1} = f(\lambda \Delta t) y_n;
\end{equation}

provided that we are dealing with explicit equations.

This function $f(\lambda \Delta t)$ determines whether our solution is stable or not. This is because for a solution to converge to a particular value, this function should be less than 1 after a certain value of n.

\begin{equation}
\vert f(\lambda \Delta t) \vert \leq 1
\end{equation}

But if we have a step function, like the runge kutta method, for a general case, when we don't know whether the method is implicit or explicit,

\begin{equation}
k1 = y_0 + h(\beta_{11} h k_1 + \beta_{12} h k_2 + \dots + \beta_{1s} h k_s)
\end{equation}
\begin{equation}
k2 = y_0 + h(\beta_{21} h k_1 + \beta_{22} h k_2 + \dots + \beta_{2s} h k_s)
\end{equation}

and so on till

\begin{equation}
ks = y_0 + h(\beta_{s1} h k_1 + \beta_{s2} h k_2 + \dots + \beta_{ss} h k_s)
\end{equation})

Let

$Y = [1 1 1 \dots 1]^T y_n$
and 
$\gamma = [\gamma_1 \gamma_2 \dots \gamma_s]^T$

where Y is a vector of size s.

We can write this in the matrix form as:

\begin{equation}
    K = Y + h \beta K
\end{equation}


\begin{equation}
    K = (1 - h \beta)^{-1}Y
\end{equation}

\begin{equation}
    y_{n+1} = y_{n} + \gamma K = y_{n} + \gamma (1 - h \beta)^{-1}Y
\end{equation}

where $K = [k_1, k_2, \dots k_s]$


\begin{equation}
    y_{n+1} = y_{n} + \gamma K = y_{n} + \gamma (1 - h \beta)^{-1}[1 1 1 \dots 1]^T y_n
\end{equation}

Let $[1 1 1 \dots 1]^T = I_{vec}$

\begin{equation}
    y_{n+1} = y_{n} + \gamma K = y_{n} + \gamma (1 - h \beta)^{-1}I_{vec} y_n = y_{n}(1 + \gamma (1 - h \beta)^{-1}I_{vec})
\end{equation}

For Runge Kutta methods to be stable, 

\begin{equation}
    (1 + \gamma (1 - h \beta)^{-1}I_{vec}) \leq 1
\end{equation}

From matrix inversion rules, we can expand $\beta$ as (if the method is explicit,only then we can apply this):

\begin{equation}
    S = 1 + h \gamma I_{vec} + h^2 \gamma \beta I_{vec}+ h^3\gamma \beta^2 I_{vec} + \dots
\end{equation}

If the scheme is a second order scheme, we can write the stability area as:

\begin{equation}
S = 1 + h\gamma I_{vec} + h^2\gamma \beta I_{vec}
\end{equation}

and if the scheme is fourth order, we can write the stability area as:

\begin{equation}
S = 1 + h\gamma I_{vec} + h^2\gamma \beta I_{vec} + h^3\gamma \beta ^2 I_{vec} + h^4 \gamma \beta^3 I_{vec}
\end{equation}

```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
#Euler Explicit
Nt = 3000;
dt = 0.005;
#Check dt = 0.1 which makes it unstable
T = [0];
dtheta = 0.1;
dv = 0.1;
theta = [sp.pi/4];
v = [0];
for i in range (0,Nt):
    theta = np.append(theta, theta[i] + v[i]*dt);
    v = np.append(v, -theta[i]*dt + v[i]);
    T = np.append(T,i*dt);

print(T);
print(v);
print(theta);
plt.plot(T,theta);
    
```

```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
#RK2
Nt = 30000;
dt = 0.001;
T = [0];
dtheta = 0.1;
dv = 0.1;
theta = [sp.pi/4];
v = [0];

def dx1dt(val1):
    dthetadt = val1;
    return dthetadt;
    
def dx2dt(val2):
    dvdt = -val2;
    return dvdt;

for i in range(0,Nt):
    k11 = dt*dx1dt(v[i]);
    k21 = dt*dx2dt(theta[i]);
    k12 = dt*dx1dt(v[i] + (2/3)*k21);
    k22 = dt*dx2dt(theta[i] + (2/3)*k11);
    theta = np.append(theta, theta[i] + (k11*0.25+k12*0.75));
    v = np.append(v, v[i]+(0.25*k21 + 0.75*k22));
    T = np.append(T,i*dt);
    
plt.plot(T,theta);
```

```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
#RK4
Nt = 3000;
dt = 0.01;
T = [0];
dtheta = 0.1;
dv = 0.1;
theta = [sp.pi/4];
v = [0];

def dx1dt(val1):
    dthetadt = val1;
    return dthetadt;
    
def dx2dt(val2):
    dvdt = -val2;
    return dvdt;

for i in range(0,Nt):
    k11 = dt*dx1dt(v[i]);
    k21 = dt*dx2dt(theta[i]);
    k12 = dt*dx1dt(v[i] + 0.5*k21);
    k22 = dt*dx2dt(theta[i] + 0.5*k11);
    k13 = dt*dx1dt(v[i] + 0.5*k22);
    k23 = dt*dx2dt(theta[i] + 0.5*k12);
    k14 = dt*dx1dt(v[i]+k23);
    k24 = dt*dx2dt(theta[i] + k23);
    theta = np.append(theta, theta[i] + (k11+2*k12+2*k13+k14)/6);
    v = np.append(v, v[i]+(k21 + 2*k22 + 2*k23 + k24)/6);
    T = np.append(T,i*dt);

plt.plot(T,theta);
```
