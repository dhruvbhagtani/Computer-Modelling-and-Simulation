Q1. Proof of Euler Maclaurin series

To prove:
\begin{equation}
    \int_a^b \frac{B_{2m+2}(\{t\})}{(2m+2)!}g^{(2)}(t)dt = \int_a^b \frac{B_{2m}(\{t\})}{(2m)!}g(t)dt + \frac{b_{2m+2}}{(2m+2)!}(g'(b)-g'(a))
\end{equation}

This integral from a to b is divided into n panels of equal size, and can be written as

\begin{equation}
    \int_a^b \frac{B_{2m+2}(\{t\})}{(2m+2)!}g^{(2)}(t)dt = \sum_{n=a}^{b-1}\int_n^{n+1} \frac{B_{2m+2}(\{t\})}{(2m+2)!}g^{(2)}(t)dt
\end{equation}

Taking one such panel, we obtain:

\begin{equation}
    \int_n^{n+1} \frac{B_{2m+2}(\{t\})}{(2m+2)!}g^{(2)}(t)dt
\end{equation}

Integrating by parts:

\begin{equation}
    \int_n^{n+1} \frac{B_{2m+2}(\{t\})}{(2m+2)!}g^{(2)}(t)dt = \frac{B_{2m+2}(t-n)}{(2m+2)!}g'(t)\Big\vert_{t=n}^{t=n+1} - \frac{1}{(2m+2)!} \int_{n}^{n+1}B'_{2m+2}(t-n)g'(t)dt
\end{equation}

Since $B'_{k}(x) = kB_{k-1}(x) $

\begin{equation}
    \int_n^{n+1} \frac{B_{2m+2}(\{t\})}{(2m+2)!}g^{(2)}(t)dt = \frac{B_{2m+2}(1)}{(2m+2)!}g'(n+1) - \frac{B_{2m+2}(1)}{(2m+2)!}g'(n) - \frac{2m+2}{(2m+2)!} \int_{n}^{n+1}B_{2m+1}(t-n)g'(t)dt
\end{equation}

Integrating by parts once again for the second term,

\begin{equation}
    \int_n^{n+1} \frac{B_{2m+2}(\{t\})}{(2m+2)!}g^{(2)}(t)dt = \frac{b_{2m+2}}{(2m+2)!}(g'(n+1) - g'(n)) - \frac{1}{(2m+1)!}\Big[B_{2m+1}(\{t\})g(t) \Big\vert_n^{n+1} -\int_n^{n+1} B'_{2m+1}(\{t\})g(t)dt \Big]
\end{equation}

Since $b_{2k+1} = 0$ for $k \geq 1$, putting limits in the second term yields zero. We are left with:

\begin{equation}
    \int_n^{n+1} \frac{B_{2m+2}(\{t\})}{(2m+2)!}g^{(2)}(t)dt = \frac{b_{2m+2}}{(2m+2)!}(g'(n+1) - g'(n)) + \frac{1}{(2m+1)!}\int_n^{n+1} B'_{2m+1}(\{t\})g(t)dt
\end{equation}

Since $B'_{k}(x) = kB_{k-1}(x) $,

\begin{equation}
    \int_n^{n+1} \frac{B_{2m+2}(\{t\})}{(2m+2)!}g^{(2)}(t)dt = \frac{b_{2m+2}}{(2m+2)!}(g'(n+1) - g'(n)) + \frac{2m+1}{(2m+1)!}\int_n^{n+1} B_{2m}(\{t\})g(t)dt
\end{equation}

This is calculated for a particular panel. If we add all the panels, the terms involing $g'(n+1) - g'(n)$ will get cancelled. We'll be left with $g'(b) - g'(a)$.


By rearranging, we get the required proof:

\begin{equation}
    \int_a^b \frac{B_{2m+2}(\{t\})}{(2m+2)!}g^{(2)}(t)dt = \int_a^b \frac{B_{2m}(\{t\})}{(2m)!}g(t)dt + \frac{b_{2m+2}}{(2m+2)!}(g'(b)-g'(a))
\end{equation}

\begin{equation}
I_n = \int_0^{\pi/2} sin^n(x)dx
\end{equation}

Integrating by parts,

\begin{equation}
I_n = \int_0^{\pi/2} sin^{n-1}(x)sin(x)dx = - sin^{n-1}(x)cos(x) + \int_{0}^{\pi/2}(n-1)sin^{n-2}(x)cos^2(x)dx
\end{equation}

Putting limits in the first term of RHS yields zero, and we are left with:

\begin{equation}
I_n = \int_{0}^{\pi/2}(n-1)sin^{n-2}(x)cos^2(x)dx = \int_{0}^{\pi/2}(n-1)sin^{n-2}(x)(1-sin^2(x))dx
\end{equation}

\begin{equation}
I_n = (n-1)I_{n-2} - (n-1)I_n
\end{equation}

\begin{equation}
nI_n = (n-1)I_{n-2}
\end{equation}

\begin{equation}
I_n = \frac{n-1}{n}I_{n-2}
\end{equation}

This is a recursive relation, and if n = 2k, we can expand it iteratively to obtain:

\begin{equation}
I_{2k} = \frac{2k-1}{2k} \frac{2k-3}{2k-2} \dots \frac{3}{4}\frac{1}{2} I_0 = \frac{2k-1}{2k} \frac{2k-3}{2k-2} \dots \frac{3}{4}\frac{1}{2} \frac{\pi}{2}
\end{equation}
Collecting 2s from denominator,

\begin{equation}
I_{2k} = \frac{1}{2^{k+1}}\frac{2k-1}{k} \frac{2k-3}{k-1} \dots \frac{3}{2}\frac{1}{1} \frac{\pi}{1}
\end{equation}

Multiply the numerator and denominator by $2^k(k+1)!$, so,

\begin{equation}
I_{2k} = \frac{\pi}{2^{2k+1}}\frac{2k-1}{k} \frac{2k-3}{k-1} \dots \frac{3}{2}\frac{1}{1} \frac{(2k)(2k-2)(2k-4) \dots 2}{k!}
\end{equation}

Rearranging,

\begin{equation}
I_{2k} = \frac{\pi}{2^{2k+1}}\frac{(2k)!}{(k)!(k)!} = \frac{\pi}{2^{2k+1}} {2k \choose k}
\end{equation}

If n = 2k+1,

\begin{equation}

\begin{equation}
I_{2k+1} = \frac{2k}{2k+1} \frac{2k-2}{2k-1} \dots \frac{4}{5}\frac{2}{3} I_1 = \frac{2k}{2k+1} \frac{2k-2}{2k-1} \dots \frac{4}{5}\frac{2}{3}

\begin{equation}
I_{2k+1} = 2^k \frac{k}{2k+1} \frac{k-1}{2k-1} \dots \frac{2}{5}\frac{1}{3} 
\end{equation}

Multiply the numerator and denominator by $2^k(k)!$

\begin{equation}
I_{2k+1} = 2^{2k+1} \frac{1}{2}\frac{k}{2k+1} \frac{k-1}{2k-1} \dots \frac{2}{5}\frac{1}{3} \frac{(k)!}{2^k(k)!}

Rearranging,

\begin{equation}
I_{2k+1} = \frac{2^{2k+1}}{2k+1} \frac{(k)!(k)!}{(2k)!}

\begin{equation}
I_{2k+1} = \frac{2^{2k+1}}{2k+1} \frac{1}{2k \choose k}

\lim_{n \to \infty} \frac{I_{2n-1}}{I_{2n+1}} = \frac{1}{4}\frac{2n+1}{2n-1}\frac{2n \choose n}{2n-2 \choose n-1}

\end{equation}

\begin{equation}
\lim_{n \to \infty} \frac{I_{2n-1}}{I_{2n+1}} = \frac{1}{4} \frac{(2n)(2n-1)}{n^2} = 1
\end{equation}

Part 3:

\begin{equation}
I_{2n} = \frac{\pi}{2^{2n+1}}\frac{(2n)!}{(n)!(n)!} = \frac{\pi}{2^{2n+1}} {2n \choose n}
\end{equation}

It is enough to prove that 

\begin{equation}
\frac{I_{2n}}{I_{2n-1}} < 1
\end{equation}
and 
\begin{equation}
\frac{I_{2n+1}}{I_{2n}} < 1
\end{equation}

For $x \in [0,\pi/2]$, sin(x) is positive. Since it is always less than 1, 

\begin{equation}
    sin^{2n-1}(x) > sin^{2n}(x) > sin^{2n+1}(x)
\end{equation}

with equality holding only at 0 and $\pi/2$.

Integrating it in limits $[0,\pi/2]$

\begin{equation}
    \int_0^{\pi/2}sin^{2n-1}(x) \geq  \int_0^{\pi/2}sin^{2n}(x) \geq  \int_0^{\pi/2}sin^{2n+1}(x)
\end{equation}

\begin{equation}
    I_{2n-1} > I_{2n} > I_{2n+1}
\end{equation}

Part 4.
By squeeze / sandwich theorem, since 

\begin{equation}
    I_{2n-1} > I_{2n} > I_{2n+1}
\end{equation}

so

\begin{equation}
    \frac{I_{2n-1}}{I_{2n+1}} > \frac{I_{2n}}{I_{2n+1}} > 1
\end{equation}

Since $\frac{I_{2n-1}}{I_{2n+1}}$ tends to infinity as n tends to infinity, 

\begin{equation}
    \lim_{n \to \infty} \frac{I_{2n}}{I_{2n+1}} = 1
\end{equation}

Part 5. 
\begin{equation}
    \frac{I_{2n}}{I_{2n+1}} = \frac{\frac{\pi}{2^{2k+1}}}{\frac{2^{2k}}{2k+1}} {2n\choose{n}}^2
\end{equation}

\begin{equation}
\frac{\pi(2k+1)}{2^{4k+1}} {2n\choose{n}}^2 = 1
end{equation}

Note this is true only for large values of n.

Since n is large, (2n+1) can be approximated as $2(n+\frac{1}{2})$, which can be written as 2(n).

Substituting, we get

\begin{equation}
{2n\choose{n}}^2 = \frac{2^{4n+1}}{\pi (2n)} = \frac{4^{2n}}{\pi n}
\end{equation}

\begin{equation}
{2n\choose{n}} = \frac{4^n}{\sqrt{n\pi}}
\end{equation}

Part 6:

We know that 

\begin{equation}
    \frac{(2n)!}{n!n!}\frac{(2n)!}{n!n!} = \frac{4^{2n}}{n \pi} 
\end{equation}

\begin{equation}
    \frac{1}{2n+1}\frac{(2n+1)!}{n!n!}\frac{(2n)!}{n!n!} = \frac{4^{2n}}{n \pi} 
\end{equation}

\begin{equation}
    \frac{1}{2n+1}\frac{(2n+1)(2n-1)(2n-3) \dots (1) 2^n n!}{n!n!}\frac{2^n n! (2n-1)(2n-3) \dots 1}{n!n!} = \frac{2^{4n}}{n \pi} 
\end{equation}

This can be written (after cancelling terms in numerator and denominator) as:

\begin{equation}
    \frac{2^{4n}}{2n+1}\frac{(2n+1)(2n-1)(2n-3) \dots (1)}{2n(2n-2) \dots 2}\frac{(2n-1)(2n-3) \dots 1}{2n(2n-21)(2n-4) \dots (2)} = \frac{2^{4n}}{n \pi} 
\end{equation}

on much simplification,

\begin{equation}
    \lim_{n \to \infty} \prod_{n=1}^\infty \Big(\frac{2k+1}{2k}\frac{2k-1}{2k} \Big) = \frac{2}{\pi}
\end{equation}

\begin{equation}
    p_n = \prod_{k=1}^n \frac{2k}{2k-1}\frac{2k}{2k+1}
\end{equation}

\begin{equation}
    p_n = \frac{1}{2n+1} \prod_{k=1}^n \frac{(2k)^4}{(2k)^2(2k-1)^2}
\end{equation}

\begin{equation}
    p_n = \frac{1}{2n+1}\frac{2^{4n}(n!)^4}{((2n)!)^2}
\end{equation}

We know that 

\begin{equation}
    n! = C \sqrt{n} (\frac{n}{e})^n
\end{equation}

Putting this value in the above equation, we obtain

\begin{equation}
    \frac{1}{2n+1}\frac{2^{4n}C^4 n^2 (\frac{n}{e})^{4n}}{C^2 2n (\frac{2n}{e})^{4n}} = \frac{\pi}{2}
\end{equation}

\begin{equation}
\frac{2n}{2(2n+1)} \frac{C^2 2^{4n}}{2^{4n}} = \pi
\end{equation}

As n goes to infinitiy, 2n is almost equal to 2n+1. With this assumption, the value of C is found as:

\begin{equation}
    C = \sqrt{2 \pi}
\end{equation}

Hence, proved.

Q3. Computing $I = \int_0^1 e^{x^2}dx$.

```python
import scipy as sp
import matplotlib.pyplot as plt
import scipy.special as spy
import numpy as np

#Midpoint rule
def Midpoint(N):
    x = sp.zeros(N);
    x = sp.arange(0,N+1)*(1/N);
    xc = sp.zeros(N);
    for j in range(0,int(N)):
        xc[j] = (x[j]+x[j+1])/2;
    f = sp.exp(xc**2);
    fsum = sum(f)*(1/N);
    return fsum;

#Trapezoidal rule
def Trapezoid(N):
    x = sp.zeros(N);
    x = sp.arange(0,N+1)*(1/N);
    f = sp.zeros(N);
    for j in range(0,int(N)):
        f[j] = (sp.exp(x[j]**2) + sp.exp(x[j+1]**2))/2;
    fsum = sum(f)*(1/N);
    return fsum;

#Trapezoidal rule with first derivatives
def TrapFD(N):
    x = sp.zeros(N);
    x = sp.arange(0,N+1)*(1/N);
    f = sp.zeros(N);
    for j in range(0,int(N)):
        f[j] = (sp.exp(x[j]**2) + sp.exp(x[j+1]**2))/2;
    fsum = sum(f)*(1/N) - (1/12)*(1/N**2)*(2*x[N]*sp.exp(x[N]) - 2*x[0]*sp.exp(x[0]));
    return fsum;

#Trapezoidal rule with first and third derivatives
def TrapTD(N):
    x = sp.zeros(N);
    x = sp.arange(0,N+1)*(1/N);
    f = sp.zeros(N);
    for j in range(0,int(N)):
        f[j] = (sp.exp(x[j]**2) + sp.exp(x[j+1]**2))/2;
    fsum = sum(f)*(1/N) - (1/12)*(1/N**2)*(2*sp.exp(1)) + (1/720)*(1/N**4)*(20*np.exp(1));
    #print(fsum);
    return fsum;

val = 11;
#val = 22;

exact = 1.46265174590718161;
emid = np.zeros(val);
etrap = np.zeros(val);
etrapFD = np.zeros(val);
etrapTD = np.zeros(val);

Num = sp.zeros(val);
for i in range(0,val):
    N = 2**i;
    Num[i] = N;
    h = 1/N;
    x = sp.zeros(N);
    x = sp.arange(0,N+1)*h;
    
    ymid = Midpoint(N);
    ytrap = Trapezoid(N);
    ytrapfd = TrapFD(N);
    ytraptd = TrapTD(N);
    emid[i] = np.abs(ymid-exact);
    etrap[i] = np.abs(ytrap-exact);
    etrapFD[i] = np.abs(ytrapfd-exact);
    etrapTD[i] = np.abs(ytraptd-exact);

plt.loglog(Num,emid,'r',label = 'Midpoint');
plt.loglog(Num,etrap,'b',label = 'Trapezoidal');
plt.loglog(Num,etrapFD,'g',label = 'Trapezoidal with first end correction');
plt.loglog(Num,etrapTD,'y',label = 'Trapezoidal with first and third end correction');
plt.legend();
plt.xlabel('Number of points');
plt.ylabel('Error between two');
plt.show();

N = 2;
i = 2
acc = 10**(-12);
error = 1;

while(abs(error)>acc):
    ymid = Midpoint(N);
    error = np.abs(ymid-exact);
    i = i+1;
    N = 2**i;   
print('No of nodes for Midpoint:')
print(N);

N = 2;
i = 2;
error = 1;

while(abs(error)>acc):
    ytrap = Trapezoid(N);
    error = np.abs(ytrap-exact);
    i = i+1;
    N = 2**i;  
print('No of nodes for Trapezoidal:')
print(N);

N = 2;
i = 2;
error = 1;

while(abs(error)>acc):
    ytrap = TrapFD(N);
    error = np.abs(ytrap-exact);
    #i = i+1;
    N = N+30;   
print('No of nodes for Trapezoidal First Correction:')
print(N);

N = 2;
i = 2;
error = 1;

while(abs(error)>acc):
    ytrap = TrapTD(N);
    error = np.abs(ytrap-exact);
    #i = i+1;
    N = N+5;  
print('No of nodes for Trapezoidal First and Third:')
print(N);
```

\begin{equation}
    P_{2n+1}(x) = \sum_{i=0}^n f(x_i)G_i(x) + \sum_{i=0}^n f'(x_i)H_i(x)
\end{equation}

If we have a point $x \in [a,b]$, which is not an interpolation point, we need to determine the error between the function $f(x)$ and its corresponding hermite interpolation. The Hermite polynomial is to be determined first.

Let the error function be a polynomial of defree 2n+2 which will describe the error between the true and the interpolated value.

\begin{equation}
    \phi(x) = P_{2n+1}(x) + \gamma \prod_{j=0}^n (x-x_j)^2
\end{equation}

This is because at all the points where the polynomial is being interpolated in the region $x \in [a,b]$, this function value should turn out to be zero. In addition, this polynomial would interpolate at $\zeta$ as well. To find the error of this new interpolant, we get,

\begin{equation}
    E(x) = f(x) - \phi(x) = f(x) - (P_{2n+1}(x) + \gamma \prod_{j=0}^n (x-x_j)^2)
\end{equation}

Since $\phi(x)$ is an interpolating function for all $x_i$ as well as $\zeta$, we obtain that the error function $E(x)$ value must be equal to zero.

\begin{equation}
 E(\zeta) = f(\zeta) - P_{2n+1}(\zeta) - \gamma \prod_{j=0}^n (\zeta -x_j)^2
\end{equation}

This is done to obtain the value of $\gamma$. So,

\begin{equation}
 \gamma = \frac{f(\zeta) - P_{2n+1}(\zeta)}{\prod_{j=0}^n (\zeta -x_j)^2}
\end{equation}

For $x_i$, since they are interpolating points, the value of $E(x_i)$ = 0.

Since hermite polynomials interpolate at first derivatives also,


\begin{equation}
 f(x_i) - P_{2n+1}(x_i) = 0
\end{equation}
\begin{equation}
 f'(x_i) - P'_{2n+1}(x_i) = 0
\end{equation}

Thus, $E(x)$ has minimum of n+2 zeros, consisting of all the interpolating points $x_i$ and $x = \zeta$. Thus, its first derivatives will have atleast n+1 zeros in $[a,b]$. Thus, the (n+2) derivative of this error function will have at least one zero. Let its value be at $x = \zeta_x$.

Differentiating the above equation for error function (2n+2) times, we obtain

\begin{equation}
    E^{2n+2}(x) = f^{2n+2}(x) - F_{2n+1}^{2n+2}(x) - \gamma \Big(\prod_{i=0}^n (x-x_i)^2 \Big)^{2n+2}
\end{equation}

The hermite polynomial is a polynomial of degree 2n+1, and vanishes at the 2n+2 derivative.

\begin{equation}
    E^{2n+2}(x) = \Big(\prod_{i=0}^n (x-x_i)^2 \Big)^{2n+2} = (2n+2)!
\end{equation}

This happens because all the terms except $x^{2m+2}$ vanish at the (2m+2) derivative. Thus, the value of $\gamma$ can be obtained as:

\begin{equation}
\gamma = \frac{f^{(2n+2)}(\zeta_x)}{(2n+2)!}
\end{equation}

Substituting this value of gamma above, we obtain,

\begin{equation}

\end{equation}

for any $\zeta$ in the domain. Replacing it by x, we get the required proof.

\begin{equation}
E(x) = f(x) - P_{2n+1}(x) = \frac{f^{(2n+2)}(\zeta_x)}{(2n+2)!} \prod_{i=0}^n (x - x_j)^2
\end{equation}

Integrating the above error term over the entire region,

\begin{equation}
\int_{-1}^1 f(x)dx - \int_{-1}^1P_{2n+1}(x) = \frac{f^{(2n+2)}(\zeta)}{(2n+2)!} \int_{-1}^1 \prod_{i=0}^n (x - x_j)^2
\end{equation}

Since these are hermite polynomials using Gaussian quadrature, the value of $P_{2m+1}(x)$ can be determined using n+1 nodal points. We can rewrite the above equation as:

\begin{equation}
\int_{-1}^1 f(x)dx - \sum_{i=0}^n w_if(x_i) = \frac{f^{(2n+2)}(\zeta)}{(2n+2)!} \int_{-1}^1 \prod_{i=0}^n (x - x_j)^2
\end{equation}

Since the nodal points are given by Gaussian quadratures with the weight function along with the first term being 1, the nodal points are nothing but the monic legendre polynomial of order (n+1).

\begin{equation}
\int_{-1}^1 f(x)dx - \sum_{i=0}^n w_if(x_i) = \frac{f^{(2n+2)}(\zeta)}{(2n+2)!} \int_{-1}^1 q_{n+1}^2(x)dx
\end{equation}

This is the required proof.
