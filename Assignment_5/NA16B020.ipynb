Q1. a. The legendre polynomial do not satisfy this recurrence relation. Instead, they satisfy the relation:

\begin{equation}
    (m+1)P_{m+1} = (2m + 1)P_m - mP_{m-1}
\end{equation}

Now, we use the generating function to prove the required relation. Also, the generating function is derived from the orthogonality principle as will be shown subsequently.

\begin{equation}
    g(t,x) = (1 - 2xt + t^2)^{-1/2} = \sum^{\infty}_{n = 0} P_n(x)t^n
\end{equation}
Differentiating the function,we get

\begin{equation}
    (-\frac{1}{2})(1 - 2xt + t^2)^{-3/2}(-2x + 2t) = (x-t)\sum^{\infty}_{n = 0} P_n(x)t^n
\end{equation}

\begin{equation}
    \frac{x-t}{(1 - 2xt + t^2)}\sum^{\infty}_{n = 0} P_n(x)t^n = (x-t)\sum^{\infty}_{n = 0} P_n(x)t^n
\end{equation}

Consider the coefficient of $t^m$, we get

\begin{equation}
    m+1P_{m+1}(x) - 2mxP_m(x) + (m-1)P_{m-1}(x) = xP_m(x) - P_{m-1}(x)
\end{equation}

\begin{equation}
    (m+1)P_{m+1}(x) = (2m+1)xP_m(x) - mP_{m-1}(x)
\end{equation}

Derivation of the generating function

In spherical coordinates, we can write as:

\begin{equation}
    \nabla^2\frac{1}{r} = \nabla^\frac{1}{z} = 0
\end{equation}

\begin{equation}
    \vert x-x_0 \vert^2 = (x-x_0)\cdot(x-x_0)
\end{equation}

\begin{equation}
    = (x\cdot x - 2x\cdot x_0 + x_0 \cdot x_0
\end{equation}

\begin{equation}
    = (r^2 - 2rcos \theta + 1)
\end{equation}


The function below is harmonic, regular at the origin and independent of $\phi$. 

\begin{equation}
    \frac{1}{(r^2 - 2rcos\theta + 1)^{0.5}}
\end{equation}

We can therefore write it in the form,

\begin{equation}
    \frac{1}{(r^2 - 2rcos\theta + 1)^{0.5}} = \sum^{\infty}_{n = 0}A_nr^nP_n(cos \theta)
\end{equation}

We have a $A_n$ term here. Let $cos \theta$ = 1, we obtain

\begin{equation}
    \frac{1}{(r^2 - 2r + 1)^{0.5}} = \frac{1}{(r-1)^2} = 1 = r+ r^2 + \dots
\end{equation}

\begin{equation}
\sum^{\infty}_{n=0}r^n = \sum^{\infty}_{n=0}A_nr^nP_n(1) = \sum^{\infty}_{n=0} A_nr^n
\end{equation}


Therefore, A_n =1. Now, expressing x as $co\theta$ and t = r, we obtain the required generating function.

\begin{equation}
g(x,t) = \frac{1}{(1-2xt + t^2)^{0.5}}
\end{equation}

which is the generating function for Legendre polynomials.

The monic legendre however, satisfy this relation. We can write as:

Inner product of $q_1(x)$ and $q_2(x)$ is given by

\begin{equation}
    \int_{a}^b q_i(x)q_j(x)w(x)dx
\end{equation}

The two functions are orthogonal. The norm of q(x) is given by

\begin{equation}
\vert \vert f(x) \vert \vert = \Big(\int_{a}^b q_i(x)q_j(x)w(x)dx \Big)
\end{equation}

 where w is the weight function.
 
 A sequence of monic polynomials witll be orthogonals if they are 0 for $ m \neq n$ and equal to $\vert \vert q_n(x) \vert \vert^2$ if the monic polynomials are $q_0(x), q_1(x), q_2(x) \dots$. This gives us legendre polynomials.
 
Also, since $q_{n+1} is a polynomial of degree n+1, and $q_n(x)$ has degree n, we can write the system as
\begin{equation}
q_{n+1}(x) - c_1q_n(x) = \sum_{k=0}^n b_kq_k(x)
\end{equation}

If we multiply it with a polynomial of degree m<n, by orthogonality principle, it must be equal to zero.

If we multiply both sides of the equation with a weight function and $q_m$(x), we get

\begin{equation}
\int^b_a q_{n+1}(x)q_m(x)w(x)dx - c_1\int_a^bxq_n(x)q_m(x)w(x)dx = \sum^n_{k=0}b_kq_k(x)q_m(x)w(x)dx
\end{equation}

For all the cases, it will be zero, except for m = n-1, and we obtain: 
\begin{equation}
    q_{n+1}(x) - c_1xP_n(x) = b_{n-1}q_{n-1}(x) + b_nq_n(x)
\end{equation}

We also know that since it is monic polynomial, c_1 has to be equal to 1, since the highest order has to be one.

Now, we have the unknows $b_{n-1}$ and b_n, which can be obtained by putting in the value of $q_0$ and $q_1$.

b. We can write the Chebyshev polynomials of first kind as:

$e^{i\theta}$ = cos\theta + isin\theta

From de Moivre theorem, we obtain
\begin{equation}
e^{in \theta} = cosn\theta + isinn\theta = (cos\theta + isin\theta)^n
\end{equation}

This can be written as:

\begin{equation}
    cosn\theta = \sum^{n}_{k=0} {{n}\choose{k}} (-1)^kcos^{n-2}\theta sin^{2k}\theta
\end{equation}

From class notes, we know tht only even powers of sin $\theta$ will appear in the expression, since the imaginary part on the RHS of the equation is zero.

Replacing the even powers of $sin\theta$ by $cos\theta$, we obtain

\begin{equation}
  T_n(x) = cos(n\theta) = \sum^{n/2}_{k=0} (-1)^k {n\choose{2k}}cos^{n-2k}\theta\Big(\sum^k_{j=0} (-1)^j {k\choose{j}}cos^{2j}\theta \Big)
\end{equation}

Assuming that n is even. If n is odd, we need to subtract n by 1 and then divide by 2 to put limits. 

After shifting the summation outside, we obtain

\begin{equation}
    cosn\theta = \sum^{n/2}_{j=0}(-1)^j\Big(\sum^{n/2}_{h=j} {n\choose{2h}}{h\choose{j}} cos^{n-2j}cos \theta \Big)
\end{equation}

To obtain the leading coefficient, we will choose j = 0 for all the cases. We will get all the values ${n\choose{0}} + {n\choose{2}} + {n\choose{4}} + {n\choose{6}} + \dots$
We know that 

${n\choose{0}} + {n\choose{1}} + {n\choose{2}} + \dots {n\choose{n}} = 2^n$

Also, choosing either all the odd terms or all the even terms is the same. Thereofore, 

\begin{equation}
{n\choose{0}} + {n\choose{2}} + {n\choose{4}} + {n\choose{6}} + \dots = \frac{2^n}{2}
\end{equation}

This gives us the desired relation.
