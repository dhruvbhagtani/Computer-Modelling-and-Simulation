3.2. Derive the following two properties:

\begin{equation}
S'_N(x_j) =

and

\begin{equation}
S''_N(x_j) = 
\begin{cases}
-\frac{\pi^2}{3h^2} - \frac{1}{6} & \text{$j \equiv 0(mod N)$} \\
\frac{(-1)^j}{2sin^2(jh)} & \text{$j \equiv 0(mod N)$}
\end{cases}
\end{equation}

Ans. 

\begin{equation}
S_N(x) = \frac{sin(\frac{\pi x}{h})}{\frac{2\pi}{h}tan(\frac{x}{2})}
\end{equation}

Differentiating wrt time, we get

\begin{equation}
S'_N(x) = \frac{cos(\frac{\pi x}{h})}{2tan(\frac{x}{2})} - \frac{h}{4\pi}sin(\frac{\pi x}{h})cosec^2(\frac{x}{2})
\end{equation}

Here, x = jh, therefore

\begin{equation}
S'_N(jh) = \frac{(-1)^j}{2tan(\frac{x}{2})}
\end{equation}

When x is near to zero, the value of tan(x/2) is close to zero, and the function is not defined there. We take the limit of the function at x = 0.

\begin{equation}
S_N'(0) = \frac{1}{x} - \frac{h}{2\pi}\frac{\pi x}{h}\frac{4}{x^2}\frac{1}{2} = 0
\end{equation}

This completes the first part of the proof.
The second part is :

\begin{equation}
S_N''(x) = -\frac{\frac{\pi}{h}sin(\frac{\pi x}{h})}{2tan(\frac{x}{2})} - \frac{cos(\frac{\pi x}{h})}{4tan^2(\frac{x}{2})}sec^2\frac{x}{2} - \frac{1}{4}cos(\frac{\pi x}{h})cosec^2(\frac{x}{2}) + \frac{h}{4 \pi}sin(\frac{\pi x}{h}) cosec^2(\frac{x}{2})cot(\frac{x}{2})
\end{equation}

When x = jh, we can write as:
\begin{equation}
S_N''(x) = \frac{(-1)^{j+1}}{2}cosec^2\frac{x}{2} - \frac{1}{4}
\end{equation}

But when x is close to zero, we need to do the taylor series approximation and find out the answer.

\begin{equation}
S_N''(x) = -\frac{\pi^2}{h^2} - \frac{2}{x^2} +  \frac{\pi^2}{h^2} + \frac{2h}{\pi x^3}(\frac{\pi x}{h} - \frac{\pi^3 x^3}{6h^3})(1-\frac{x^2}{4})
\end{equation}

\begin{equation}
S_N''(x) = -\frac{\pi^2}{3h^2} - \frac{1}{6}
\end{equation}

Ans.

To obtain the third order spectral differential method, we need to differentiate $S_N(x)$ three times. From previous question, we have 

\begin{equation}
S_N''(x) = -\frac{\frac{\pi}{h}sin(\frac{\pi x}{h})}{2tan(\frac{x}{2})} - \frac{cos(\frac{\pi x}{h})}{2sin^2(\frac{x}{2})} + \frac{h}{4 \pi}sin(\frac{\pi x}{h}) cosec^2(\frac{x}{2})cot(\frac{x}{2})
\end{equation}

Either we can differentiate this, or we can find out the value of the integral

\begin{equation}
a_j = \frac{1}{2\pi}\int_{-\pi/h}^{\pi/h}(i\zeta)^3e^{i\zeta jh}d\zeta
\end{equation}

It is to note that the terms involving sine will turn out to be zero under the limits, and are therefore, ignored in the calculations.

\begin{equation}
a_j = \frac{-1}{2\pi}\int_{-\pi/h}^{\pi/h}(\zeta)^3e^{i\zeta jh}d\zeta
\end{equation}

\begin{equation}
a_j = \frac{-i}{2\pi}\Big[\frac{\zeta^3}{ijh}e^{ij\zeta h}\Big]^{\pi/h}_{-\pi/h} + \frac{i}{2\pi}\int_{-\pi/h}^{\pi/h}(3\zeta)^2\frac{1}{ijh}e^{i\zeta jh}d\zeta
\end{equation}

Now, doing integration by parts again for the second term, we get

\begin{equation}
a_j = \frac{-1}{h^2 j \pi}\frac{\pi^3}{h^3}cos(j\pi) + sine term - \frac{3}{2hj\pi}\int_{-\pi/h}^{\pi/h}2\zeta\frac{1}{ijh}e^{i\zeta jh}d\zeta
\end{equation}

\begin{equation}
a_j = \frac{\pi^2}{jh^4}(-1)^{j+1} - \frac{3}{h^2j^2\pi i}\int_{-\pi/h}^{\pi/h}\zeta e^{i\zeta jh}d\zeta
\end{equation}

\begin{equation}
a_j = \frac{\pi^2}{jh^4}(-1)^{j+1} + \frac{6}{h^3 j^3}(-1)^j
\end{equation}

However, this is valid only for j \neq 0. When j = 0, then we put j = 0 in the first equation and the integral turns out to be zero. So,

$a_j$ = 0 when $j = 0$

3.8 Recompute output 6 by a modified program using leap frog formula rather than a spectral method. Produce plots for N = 128 and N = 256. Comment.

```python
# SECOND ORDER FINITE DIFFERENCE WITH LEAP FROG
# Grid, variable coefficient and initial data
import scipy as sp;
from scipy import linalg;
from scipy import sparse;
pi = sp.pi;
N  = 128;      # Number of grid points
h  = 2*pi/N; # Grid spacing
print(h)         # Display grid spacing
x  = h*sp.arange(1,N+1); # Grid points
v           = sp.zeros(N);
v[1]        = -0.5/h;
v[-1]       = 0.5/h;
D           = sparse.csr_matrix(sp.linalg.circulant(v));

tinit           = 0;         # Intial time
tmax         = 8;         # Maximum time
nsteps       = 800;    # Number of steps
dt              = (tmax-tinit)/nsteps;  # time step
print(dt)           # Display time step
#c = 0.2; # Constant Wave speed
c = 0.2+sp.sin(x-1)**2; # Variable Wave speed
v   = sp.exp(-100*(x-1)**2); # Initial condition
vold = sp.exp(-100*(x-c*dt-1)**2); # Second initial condition for leap frod
# plotstep = 10; # How frequent should I plot?
# tplot      = plotstep*dt;
# nplots    = nsteps//plotstep;
data = sp.zeros((nsteps+1,N));
data[0,:] = v;
tdata = sp.zeros(nsteps+1);
tdata[0] = tinit;
for i in sp.arange(0,nsteps):
    vnew = vold-2*dt*c*(D.dot(v));
    vold = v;
    v = vnew;
    data[i+1,:] = v;
    tdata[i+1] = tdata[i]+dt;

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np


X, T = np.meshgrid(x, tdata, sparse=False, indexing='ij')

fig = plt.figure();
ax = fig.add_subplot(111, projection='3d');
# X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_wireframe(X.T, T.T, data, rstride=20, cstride=20)
ax.view_init(60,-75)
plt.show();

plt.pcolor(x, tdata, data)
plt.colorbar()
plt.title('Second order with Leap Frog')
plt.show();
```
